```python
ihsg_monthly_df['year_month'] = ihsg_monthly_df['Tanggal'].dt.to_period('M')
model1_df['year_month'] = model1_df['date'].dt.to_period('M')

model1_df_merged = model1_df.merge(
    ihsg_monthly_df[['year_month', 'Terakhir', 'Pembukaan', 'Tertinggi', 'Terendah', 'Perubahan%']], 
    on='year_month', 
    how='left',
    suffixes=('', '_ihsg')
)

model1_df_merged = model1_df_merged.rename(columns={
    'Terakhir': 'ihsg_close',
    'Pembukaan': 'ihsg_open',
    'Tertinggi': 'ihsg_high',
    'Terendah': 'ihsg_low',
    'Perubahan%': 'ihsg_return'
})
```

```python
model1_df_merged = model1_df_merged.drop('year_month', axis=1)
```

```python
model1_df_cleaned = model1_df_merged.copy()
```

```python
model1_df_cleaned = model1_df_cleaned.drop(['ihsg_open', 'ihsg_high', 'ihsg_low'], axis=1)
```

```python
model1_df_cleaned
```

```python
model1_df_cleaned.info()
```

```python
prices_columns = ['usd_idr_close', 'usd_idr_open', 'usd_idr_high', 'usd_idr_low', 'idr_m1', 'idr_m2', 'total_credit', 'non_performing_loan', 'id_leading_economic_indicator', 'ihsg_close']
ratio_columns = ['usd_idr_returns', 'id_inflation_rate_mom', 'id_inflation_rate_yoy', 'bi_interest_rate', 'npl_ratio',  'ihsg_return']
```

```python
from sklearn.preprocessing import StandardScaler

for col in prices_columns:
    if model1_df_cleaned[col].dtype == 'object':
        model1_df_cleaned[col] = model1_df_cleaned[col].str.replace(',', '', regex=False).astype(float)
        
    scaler = StandardScaler()
    model1_df_cleaned[[col]] = scaler.fit_transform(model1_df_cleaned[[col]])
```

```python
model1_df_cleaned
```

```python
for col in ratio_columns:
    if model1_df_cleaned[col].dtype == 'object':
        model1_df_cleaned[col] = model1_df_cleaned[col].str.replace('%', '').astype(float) / 100.0
    else:
        model1_df_cleaned[col] = model1_df_cleaned[col].astype(float) / 100.0

    scaler = StandardScaler()
    model1_df_cleaned[[col]] = scaler.fit_transform(model1_df_cleaned[[col]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

COLORS = {
    'IHSG': '#1f77b4',
    'Inflation': '#d62728',
    'BI_Rate': '#ff7f0e',
    'Credit': '#2ca02c',
    'Money_Supply': '#17becf',
    'LEI': "#62447e",
    'USD_IDR': '#8c564b',
    'NPL': '#e377c2',
    'Net_Buy': '#2ca02c',
}

def setup_thesis_style():
    plt.style.use('default')
    sns.set_style("ticks")
    sns.set_palette(list(COLORS.values()))
    plt.rcParams.update({
        'font.family': 'Gill Sans',
        'font.size': 12,
        'axes.titlesize': 16,
        'axes.labelsize': 14,
        'axes.titleweight': 'light',
        'axes.grid': True,
        'grid.alpha': 0.2,
        'grid.linestyle': '--',
        'grid.color': '#aaaaaa',
        'xtick.labelsize': 12,
        'ytick.labelsize': 12,
        'legend.fontsize': 12,
        'figure.titlesize': 18
    })

def add_footnote(ax, note="Source: Author’s calculation, 2025"):
    ax.text(1, -0.15, note,
            transform=ax.transAxes,
            ha='right', va='top', fontsize=10, color='gray')
    

def datalabel_bar(ax, fontsize=12):
    for rect in ax.patches:
        height = rect.get_height()
        if height == 0:
            continue
        offset = 3 if height > 0 else -15
        ax.annotate(f"{height:.2f}",
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, offset),
                    textcoords="offset points",
                    ha="center",
                    va="bottom" if height > 0 else "top",
                    fontsize=fontsize, color='dimgray')

def adjust_xticks(ax, rotation=30):
    for label in ax.get_xticklabels():
        label.set_rotation(rotation)
        label.set_horizontalalignment('right')

def finalize_plot(ax, title=None, xlabel=None, ylabel=None, legend_loc='best'):
    ax.set_title(title or "")
    ax.set_xlabel(xlabel or "")
    ax.set_ylabel(ylabel or "")
    ax.legend(loc=legend_loc, frameon=False)
    adjust_xticks(ax)
    plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8,5))

sns.lineplot(data=ihsg_to_netbuying_df, x='Date', y='IHSG',
             ax=axes, color=COLORS['IHSG'], label='IHSG',
             marker='o', markersize=3, alpha=0.9)
sns.lineplot(data=ihsg_to_netbuying_df, x='Date', y='Foreign Net Buying',
             ax=axes, color=COLORS['Net_Buy'], label='Foreign Net Buying',
             marker='o', markersize=3, alpha=0.9)
axes.grid(True, alpha=0.6, linestyle='-', linewidth=1.0, axis='y')
finalize_plot(axes, 
              title='IHSG and Foreign Net Buying',
              ylabel='Standardized Value (Z-Score)',
              legend_loc='upper left')
plt.show()
```
