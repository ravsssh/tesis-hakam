````python
# Konversi format desimal untuk kolom numerik
def convert_decimal_format(df, columns):
    for col in columns:
        if col in df.columns:
            # Jika kolom berisi string, konversi format desimal
            if df[col].dtype == 'object':
                df[col] = df[col].str.replace('.', '').str.replace(',', '.').astype(float)
            # Jika sudah dalam format numerik tapi menggunakan locale yang berbeda
            elif df[col].dtype in ['int64', 'float64']:
                pass  # Sudah dalam format yang benar
    return df
````

````python
# ðŸ“Š PEARSON CORRELATION ANALYSIS WITH DATE ALIGNMENT
print("ðŸ”— ALIGNING DATES AND CALCULATING PEARSON CORRELATIONS")
print("="*60)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Step 1: Prepare dataframes with consistent date columns
print("Step 1: Preparing dataframes...")

# Create copies and ensure date columns are datetime
ihsg_corr = ihsg_df.copy()
sti_corr = sti_df.copy()
xauusd_corr = xauusd_df.copy()

# Ensure date columns are datetime
ihsg_corr['Date'] = pd.to_datetime(ihsg_corr['Date'])
sti_corr['Date'] = pd.to_datetime(sti_corr['Date'])  
xauusd_corr['Tanggal'] = pd.to_datetime(xauusd_corr['Tanggal'])

# Rename xauusd date column for consistency
xauusd_corr = xauusd_corr.rename(columns={'Tanggal': 'Date'})

print(f"IHSG date range: {ihsg_corr['Date'].min()} to {ihsg_corr['Date'].max()}")
print(f"STI date range: {sti_corr['Date'].min()} to {sti_corr['Date'].max()}")
print(f"XAUUSD date range: {xauusd_corr['Date'].min()} to {xauusd_corr['Date'].max()}")

# Step 2: Find common dates across all three dataframes
print("\nStep 2: Finding common dates...")

ihsg_dates = set(ihsg_corr['Date'])
sti_dates = set(sti_corr['Date'])
xauusd_dates = set(xauusd_corr['Date'])

# Find intersection of all three date sets
common_dates = ihsg_dates.intersection(sti_dates).intersection(xauusd_dates)
common_dates = sorted(list(common_dates))

print(f"Total common dates: {len(common_dates)}")
print(f"Common date range: {min(common_dates)} to {max(common_dates)}")

# Step 3: Filter dataframes to only include common dates
print("\nStep 3: Filtering to common dates...")

ihsg_aligned = ihsg_corr[ihsg_corr['Date'].isin(common_dates)].copy()
sti_aligned = sti_corr[sti_corr['Date'].isin(common_dates)].copy()
xauusd_aligned = xauusd_corr[xauusd_corr['Date'].isin(common_dates)].copy()

# Sort by date
ihsg_aligned = ihsg_aligned.sort_values('Date').reset_index(drop=True)
sti_aligned = sti_aligned.sort_values('Date').reset_index(drop=True)
xauusd_aligned = xauusd_aligned.sort_values('Date').reset_index(drop=True)

print(f"IHSG aligned samples: {len(ihsg_aligned)}")
print(f"STI aligned samples: {len(sti_aligned)}")
print(f"XAUUSD aligned samples: {len(xauusd_aligned)}")

# Step 4: Create correlation dataframe
print("\nStep 4: Creating correlation dataset...")

# Merge all three datasets on Date
correlation_df = pd.merge(
    ihsg_aligned[['Date', 'Close']].rename(columns={'Close': 'IHSG_Close'}),
    sti_aligned[['Date', 'Close']].rename(columns={'Close': 'STI_Close'}),
    on='Date',
    how='inner'
)

correlation_df = pd.merge(
    correlation_df,
    xauusd_aligned[['Date', 'Terakhir']].rename(columns={'Terakhir': 'XAUUSD_Close'}),
    on='Date',
    how='inner'
)

# Remove any rows with missing values
correlation_df = correlation_df.dropna()

print(f"Final correlation dataset: {len(correlation_df)} samples")
print(f"Date range: {correlation_df['Date'].min()} to {correlation_df['Date'].max()}")

# Step 5: Calculate Pearson correlations
print(f"\nðŸ“Š PEARSON CORRELATION RESULTS:")
print("-"*50)

# Extract price columns for correlation
prices = correlation_df[['IHSG_Close', 'STI_Close', 'XAUUSD_Close']]

# Calculate correlation matrix
correlation_matrix = prices.corr(method='pearson')

print("Correlation Matrix:")
print(correlation_matrix.round(4))

# Calculate individual correlations with p-values
correlations = {}

# IHSG vs STI
corr_ihsg_sti, p_val_ihsg_sti = pearsonr(correlation_df['IHSG_Close'], correlation_df['STI_Close'])
correlations['IHSG_vs_STI'] = {'correlation': corr_ihsg_sti, 'p_value': p_val_ihsg_sti}

# IHSG vs XAUUSD
corr_ihsg_xau, p_val_ihsg_xau = pearsonr(correlation_df['IHSG_Close'], correlation_df['XAUUSD_Close'])
correlations['IHSG_vs_XAUUSD'] = {'correlation': corr_ihsg_xau, 'p_value': p_val_ihsg_xau}

# STI vs XAUUSD
corr_sti_xau, p_val_sti_xau = pearsonr(correlation_df['STI_Close'], correlation_df['XAUUSD_Close'])
correlations['STI_vs_XAUUSD'] = {'correlation': corr_sti_xau, 'p_value': p_val_sti_xau}

print(f"\nDetailed Correlation Results:")
print("-"*40)
for pair, stats in correlations.items():
    significance = "***" if stats['p_value'] < 0.001 else "**" if stats['p_value'] < 0.01 else "*" if stats['p_value'] < 0.05 else ""
    strength = "Very Strong" if abs(stats['correlation']) > 0.8 else "Strong" if abs(stats['correlation']) > 0.6 else "Moderate" if abs(stats['correlation']) > 0.4 else "Weak" if abs(stats['correlation']) > 0.2 else "Very Weak"
    
    print(f"{pair}:")
    print(f"  â€¢ Correlation: {stats['correlation']:.4f}{significance}")
    print(f"  â€¢ P-value: {stats['p_value']:.6f}")
    print(f"  â€¢ Strength: {strength}")
    print()

# Step 6: Visualizations
print("Step 6: Creating correlation visualizations...")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Pearson Correlation Analysis - Aligned Dataset', fontsize=16, fontweight='bold')

# 1. Correlation Heatmap
ax1 = axes[0, 0]
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', center=0, 
           square=True, ax=ax1, fmt='.4f', cbar_kws={'shrink': 0.8})
ax1.set_title('Correlation Matrix Heatmap')

# 2. IHSG vs STI Scatter
ax2 = axes[0, 1]
ax2.scatter(correlation_df['STI_Close'], correlation_df['IHSG_Close'], alpha=0.6, color='blue')
ax2.set_xlabel('STI Close Price')
ax2.set_ylabel('IHSG Close Price')
ax2.set_title(f'IHSG vs STI\n(r = {corr_ihsg_sti:.4f}, p = {p_val_ihsg_sti:.4f})')
ax2.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(correlation_df['STI_Close'], correlation_df['IHSG_Close'], 1)
p = np.poly1d(z)
ax2.plot(correlation_df['STI_Close'], p(correlation_df['STI_Close']), "r--", alpha=0.8)

# 3. IHSG vs XAUUSD Scatter
ax3 = axes[0, 2]
ax3.scatter(correlation_df['XAUUSD_Close'], correlation_df['IHSG_Close'], alpha=0.6, color='green')
ax3.set_xlabel('XAUUSD Close Price')
ax3.set_ylabel('IHSG Close Price')
ax3.set_title(f'IHSG vs XAUUSD\n(r = {corr_ihsg_xau:.4f}, p = {p_val_ihsg_xau:.4f})')
ax3.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(correlation_df['XAUUSD_Close'], correlation_df['IHSG_Close'], 1)
p = np.poly1d(z)
ax3.plot(correlation_df['XAUUSD_Close'], p(correlation_df['XAUUSD_Close']), "r--", alpha=0.8)

# 4. STI vs XAUUSD Scatter
ax4 = axes[1, 0]
ax4.scatter(correlation_df['XAUUSD_Close'], correlation_df['STI_Close'], alpha=0.6, color='orange')
ax4.set_xlabel('XAUUSD Close Price')
ax4.set_ylabel('STI Close Price')
ax4.set_title(f'STI vs XAUUSD\n(r = {corr_sti_xau:.4f}, p = {p_val_sti_xau:.4f})')
ax4.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(correlation_df['XAUUSD_Close'], correlation_df['STI_Close'], 1)
p = np.poly1d(z)
ax4.plot(correlation_df['XAUUSD_Close'], p(correlation_df['XAUUSD_Close']), "r--", alpha=0.8)

# 5. Time Series Plot
ax5 = axes[1, 1]
# Normalize for better visualization
ihsg_norm = (correlation_df['IHSG_Close'] / correlation_df['IHSG_Close'].iloc[0]) * 100
sti_norm = (correlation_df['STI_Close'] / correlation_df['STI_Close'].iloc[0]) * 100
xauusd_norm = (correlation_df['XAUUSD_Close'] / correlation_df['XAUUSD_Close'].iloc[0]) * 100

ax5.plot(correlation_df['Date'], ihsg_norm, label='IHSG (Normalized)', alpha=0.8)
ax5.plot(correlation_df['Date'], sti_norm, label='STI (Normalized)', alpha=0.8)
ax5.plot(correlation_df['Date'], xauusd_norm, label='XAUUSD (Normalized)', alpha=0.8)
ax5.set_title('Normalized Price Movement (Base=100)')
ax5.set_ylabel('Normalized Price')
ax5.legend()
ax5.grid(True, alpha=0.3)
ax5.tick_params(axis='x', rotation=45)

# 6. Rolling Correlation
ax6 = axes[1, 2]
window_size = 252  # 1 year rolling window

rolling_corr_ihsg_sti = correlation_df['IHSG_Close'].rolling(window=window_size).corr(correlation_df['STI_Close'])
rolling_corr_ihsg_xau = correlation_df['IHSG_Close'].rolling(window=window_size).corr(correlation_df['XAUUSD_Close'])

ax6.plot(correlation_df['Date'], rolling_corr_ihsg_sti, label='IHSG-STI', alpha=0.8)
ax6.plot(correlation_df['Date'], rolling_corr_ihsg_xau, label='IHSG-XAUUSD', alpha=0.8)
ax6.set_title(f'Rolling Correlation ({window_size}-day window)')
ax6.set_ylabel('Correlation Coefficient')
ax6.legend()
ax6.grid(True, alpha=0.3)
ax6.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Step 7: Summary statistics
print(f"\nðŸ“ˆ SUMMARY STATISTICS FOR ALIGNED DATASET:")
print("-"*50)
summary_stats = correlation_df[['IHSG_Close', 'STI_Close', 'XAUUSD_Close']].describe()
print(summary_stats.round(2))

# Check for data quality
print(f"\nðŸ” DATA QUALITY CHECK:")
print("-"*30)
print(f"Missing values:")
print(correlation_df[['IHSG_Close', 'STI_Close', 'XAUUSD_Close']].isnull().sum())
print(f"\nData types:")
print(correlation_df[['IHSG_Close', 'STI_Close', 'XAUUSD_Close']].dtypes)

# Save aligned dataset
correlation_df.to_csv('aligned_correlation_dataset.csv', index=False)
print(f"\nðŸ’¾ Aligned dataset saved as 'aligned_correlation_dataset.csv'")

print(f"\nðŸŽ¯ KEY INSIGHTS:")
print("-"*30)
print(f"â€¢ Dataset contains {len(correlation_df):,} aligned observations")
print(f"â€¢ Strongest correlation: {max(abs(corr_ihsg_sti), abs(corr_ihsg_xau), abs(corr_sti_xau)):.4f}")
print(f"â€¢ All correlations are statistically significant at p < 0.05" if all(p < 0.05 for p in [p_val_ihsg_sti, p_val_ihsg_xau, p_val_sti_xau]) else "â€¢ Some correlations may not be statistically significant")
print("="*60)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from darts import TimeSeries
from darts.dataprocessing.transformers import MissingValuesFiller
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Assuming you have already created your TimeSeries objects:
# ihsg_series = TimeSeries.from_dataframe(ihsg_df, time_col='Date', value_cols=['Open'], fill_missing_dates=True, freq='B')
# sti_series = TimeSeries.from_dataframe(sti_df, time_col='Date', value_cols=['Open'], fill_missing_dates=True, freq='B')
# xauusd_series = TimeSeries.from_dataframe(xauusd_df, time_col='Tanggal', value_cols=['Pembukaan'], fill_missing_dates=True, freq='B')

def calculate_descriptive_stats(series, name):
    """Calculate comprehensive descriptive statistics for a time series"""
    values = series.values().flatten()
    
    stats = {
        'Series': name,
        'Count': len(values),
        'Mean': np.mean(values),
        'Median': np.median(values),
        'Std Dev': np.std(values),
        'Min': np.min(values),
        'Max': np.max(values),
        'Range': np.max(values) - np.min(values),
        'Skewness': pd.Series(values).skew(),
        'Kurtosis': pd.Series(values).kurtosis(),
        'Q1 (25%)': np.percentile(values, 25),
        'Q3 (75%)': np.percentile(values, 75),
        'IQR': np.percentile(values, 75) - np.percentile(values, 25),
        'CV (%)': (np.std(values) / np.mean(values)) * 100,  # Coefficient of Variation
        'Start Date': series.start_time(),
        'End Date': series.end_time(),
        'Missing Values': series.gaps().sum() if hasattr(series, 'gaps') else 0
    }
    
    return stats

def create_eda_plots(series_dict):
    """Create comprehensive EDA plots for multiple time series"""
    
    # Create a large figure with multiple subplots
    fig = plt.figure(figsize=(20, 16))
    
    # 1. Time Series Line Plot
    ax1 = plt.subplot(3, 3, 1)
    for name, series in series_dict.items():
        series.plot(ax=ax1, label=name, linewidth=1.5)
    ax1.set_title('Time Series Overview', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Price')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Normalized Time Series (for comparison)
    ax2 = plt.subplot(3, 3, 2)
    for name, series in series_dict.items():
        # Normalize to start from 100
        normalized = (series / series.values()[0][0]) * 100
        normalized.plot(ax=ax2, label=name, linewidth=1.5)
    ax2.set_title('Normalized Time Series (Base=100)', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Normalized Price')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Distribution plots (histograms)
    ax3 = plt.subplot(3, 3, 3)
    for name, series in series_dict.items():
        values = series.values().flatten()
        ax3.hist(values, alpha=0.6, bins=30, label=name, density=True)
    ax3.set_title('Price Distribution', fontsize=14, fontweight='bold')
    ax3.set_xlabel('Price')
    ax3.set_ylabel('Density')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 4. Box plots
    ax4 = plt.subplot(3, 3, 4)
    box_data = [series.values().flatten() for series in series_dict.values()]
    box_labels = list(series_dict.keys())
    ax4.boxplot(box_data, labels=box_labels)
    ax4.set_title('Box Plot Comparison', fontsize=14, fontweight='bold')
    ax4.set_ylabel('Price')
    ax4.grid(True, alpha=0.3)
    plt.setp(ax4.get_xticklabels(), rotation=45)
    
    # 5. Returns calculation and plot
    ax5 = plt.subplot(3, 3, 5)
    for name, series in series_dict.items():
        returns = series.diff().dropna()
        returns.plot(ax=ax5, label=f'{name} Returns', linewidth=0.8, alpha=0.8)
    ax5.set_title('Daily Returns', fontsize=14, fontweight='bold')
    ax5.set_ylabel('Returns')
    ax5.legend()
    ax5.grid(True, alpha=0.3)
    
    # 6. Returns distribution
    ax6 = plt.subplot(3, 3, 6)
    for name, series in series_dict.items():
        returns = series.diff().dropna().values().flatten()
        ax6.hist(returns, alpha=0.6, bins=40, label=f'{name} Returns', density=True)
    ax6.set_title('Returns Distribution', fontsize=14, fontweight='bold')
    ax6.set_xlabel('Returns')
    ax6.set_ylabel('Density')
    ax6.legend()
    ax6.grid(True, alpha=0.3)
    
    # 7. Rolling statistics (30-day window)
    ax7 = plt.subplot(3, 3, 7)
    for name, series in series_dict.items():
        rolling_mean = series.pd_dataframe().rolling(window=30).mean()
        rolling_std = series.pd_dataframe().rolling(window=30).std()
        ax7.plot(rolling_mean.index, rolling_mean.values, label=f'{name} 30d Mean', linewidth=1.5)
    ax7.set_title('30-Day Rolling Mean', fontsize=14, fontweight='bold')
    ax7.set_ylabel('Price')
    ax7.legend()
    ax7.grid(True, alpha=0.3)
    
    # 8. Volatility (30-day rolling std)
    ax8 = plt.subplot(3, 3, 8)
    for name, series in series_dict.items():
        rolling_std = series.pd_dataframe().rolling(window=30).std()
        ax8.plot(rolling_std.index, rolling_std.values, label=f'{name} Volatility', linewidth=1.5)
    ax8.set_title('30-Day Rolling Volatility', fontsize=14, fontweight='bold')
    ax8.set_ylabel('Standard Deviation')
    ax8.legend()
    ax8.grid(True, alpha=0.3)
    
    # 9. Correlation heatmap (if multiple series)
    ax9 = plt.subplot(3, 3, 9)
    if len(series_dict) > 1:
        # Create correlation matrix
        corr_data = pd.DataFrame()
        for name, series in series_dict.items():
            corr_data[name] = series.pd_dataframe().values.flatten()
        
        correlation_matrix = corr_data.corr()
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
                   square=True, ax=ax9, fmt='.3f')
        ax9.set_title('Correlation Matrix', fontsize=14, fontweight='bold')
    else:
        ax9.text(0.5, 0.5, 'Correlation requires\nmultiple series', 
                ha='center', va='center', transform=ax9.transAxes, fontsize=12)
        ax9.set_title('Correlation Matrix', fontsize=14, fontweight='bold')
    
    plt.tight_layout(pad=3.0)
    plt.show()

def print_summary_stats(series_dict):
    """Print comprehensive summary statistics"""
    print("="*80)
    print("TIME SERIES DESCRIPTIVE STATISTICS")
    print("="*80)
    
    stats_list = []
    for name, series in series_dict.items():
        stats = calculate_descriptive_stats(series, name)
        stats_list.append(stats)
    
    stats_df = pd.DataFrame(stats_list)
    
    # Print basic statistics
    print("\nðŸ“Š BASIC STATISTICS")
    print("-"*80)
    basic_cols = ['Series', 'Count', 'Mean', 'Median', 'Std Dev', 'Min', 'Max']
    print(stats_df[basic_cols].to_string(index=False, float_format='%.2f'))
    
    # Print distribution statistics
    print("\nðŸ“ˆ DISTRIBUTION STATISTICS")
    print("-"*80)
    dist_cols = ['Series', 'Skewness', 'Kurtosis', 'Q1 (25%)', 'Q3 (75%)', 'IQR', 'CV (%)']
    print(stats_df[dist_cols].to_string(index=False, float_format='%.3f'))
    
    # Print time range information
    print("\nðŸ“… TIME RANGE INFORMATION")
    print("-"*80)
    time_cols = ['Series', 'Start Date', 'End Date', 'Missing Values']
    print(stats_df[time_cols].to_string(index=False))

def calculate_returns_stats(series_dict):
    """Calculate and display returns statistics"""
    print("\nðŸ“Š RETURNS STATISTICS")
    print("-"*80)
    
    returns_stats = []
    for name, series in series_dict.items():
        returns = series.diff().dropna().values().flatten()
        
        returns_stat = {
            'Series': name,
            'Mean Return': np.mean(returns),
            'Std Return': np.std(returns),
            'Min Return': np.min(returns),
            'Max Return': np.max(returns),
            'Sharpe Ratio': np.mean(returns) / np.std(returns) * np.sqrt(252),  # Annualized
            'Positive Days (%)': (np.sum(returns > 0) / len(returns)) * 100,
            'Volatility (Ann.)': np.std(returns) * np.sqrt(252) * 100
        }
        returns_stats.append(returns_stat)
    
    returns_df = pd.DataFrame(returns_stats)
    print(returns_df.to_string(index=False, float_format='%.4f'))

# Example usage:
# Create a dictionary with your series
series_dict = {
    'IHSG': ihsg_series,
    'STI': sti_series,
    'Gold (XAU/USD)': xauusd_series
}

# Run the EDA
print_summary_stats(series_dict)
calculate_returns_stats(series_dict)
create_eda_plots(series_dict)

# Additional analysis: Correlation between series (if you have multiple)
print("\nðŸ” ADDITIONAL INSIGHTS")
print("-"*80)

for name, series in series_dict.items():
    values = series.values().flatten()
    print(f"\n{name}:")
    print(f"  â€¢ Average daily change: {np.mean(np.diff(values)):.2f}")
    print(f"  â€¢ Maximum single-day gain: {np.max(np.diff(values)):.2f}")
    print(f"  â€¢ Maximum single-day loss: {np.min(np.diff(values)):.2f}")
    print(f"  â€¢ Days with positive returns: {np.sum(np.diff(values) > 0)} out of {len(np.diff(values))}")
    
    # Calculate trend (simple linear regression slope)
    x = np.arange(len(values))
    slope = np.polyfit(x, values, 1)[0]
    if slope > 0:
        trend = "ðŸ“ˆ Upward"
    elif slope < 0:
        trend = "ðŸ“‰ Downward"
    else:
        trend = "âž¡ï¸ Sideways"
    print(f"  â€¢ Overall trend: {trend} (slope: {slope:.4f})")
````

````python
# Find common time period for all series
common_start = max(ihsg_series.start_time(), sti_series.start_time(), xauusd_series.start_time())
common_end = min(ihsg_series.end_time(), sti_series.end_time(), xauusd_series.end_time())

print(f"Common time period: {common_start} to {common_end}")

# Slice all series to the common period
ihsg_common = ihsg_series.slice(common_start, common_end)
sti_common = sti_series.slice(common_start, common_end)
xauusd_common = xauusd_series.slice(common_start, common_end)

# Fill missing values using auto method
filler = MissingValuesFiller(fill="auto")

ihsg_common = filler.transform(ihsg_common)
sti_common = filler.transform(sti_common)
xauusd_common = filler.transform(xauusd_common)

print(f"\nAligned series lengths:")
print(f"IHSG: {len(ihsg_common)}")
print(f"STI: {len(sti_common)}")
print(f"XAUUSD: {len(xauusd_common)}")

print(f"\nFinal aligned lengths:")
print(f"IHSG (target): {len(ihsg_common)}")
print(f"STI (covariate): {len(sti_common)}")
print(f"XAUUSD (covariate): {len(xauusd_common)}")
````

````python
# Split data for training and testing
# Use the last year for training and forecast 1 month ahead

# Calculate split point: keep last 365 days for training
train_days = 365  # 1 year
forecast_horizon = 30  # 1 month

if len(ihsg_common) > train_days + forecast_horizon:
    # Split data
    train_end_idx = len(ihsg_common) - forecast_horizon
    train_start_idx = train_end_idx - train_days
    
    # Ensure we have enough data
    if train_start_idx < 0:
        train_start_idx = 0
    
    # Create training sets
    ihsg_train = ihsg_common[train_start_idx:train_end_idx]
    sti_train = sti_common[train_start_idx:train_end_idx]
    xauusd_train = xauusd_common[train_start_idx:train_end_idx]
    
    # Create test sets (for validation)
    ihsg_test = ihsg_common[train_end_idx:]
    sti_test = sti_common[train_end_idx:]
    xauusd_test = xauusd_common[train_end_idx:]
    
    print(f"Training period: {ihsg_train.start_time()} to {ihsg_train.end_time()}")
    print(f"Test period: {ihsg_test.start_time()} to {ihsg_test.end_time()}")
    print(f"Training samples: {len(ihsg_train)}")
    print(f"Test samples: {len(ihsg_test)}")
    
else:
    # If not enough data, use most recent data for training
    ihsg_train = ihsg_common[:-forecast_horizon] if len(ihsg_common) > forecast_horizon else ihsg_common
    sti_train = sti_common[:-forecast_horizon] if len(sti_common) > forecast_horizon else sti_common
    xauusd_train = xauusd_common[:-forecast_horizon] if len(xauusd_common) > forecast_horizon else xauusd_common
    
    print(f"Using available data for training: {len(ihsg_train)} samples")
    print(f"Training period: {ihsg_train.start_time()} to {ihsg_train.end_time()}")
````

# ðŸ”¢ Mathematical Formulation & Technical Details

## Mathematical Model

### Target Variable
**IHSG(t)** = Closing price of Indonesia Stock Exchange Composite Index at time t

### Covariates
- **STI(t)** = Singapore Straits Times Index at time t
- **XAUUSD(t)** = Gold price in USD at time t

### Feature Vector Construction
For prediction at time **t**, the feature vector **X(t)** is constructed as:

```
X(t) = [
    IHSG(t-1), IHSG(t-2), ..., IHSG(t-20),     # Target lags
    STI(t-1), STI(t-2), ..., STI(t-20),        # STI lags  
    XAUUSD(t-1), XAUUSD(t-2), ..., XAUUSD(t-20) # XAUUSD lags
]
```

**Dimension**: X(t) âˆˆ â„â¶â°

### Random Forest Prediction
The Random Forest model consists of **B = 100** decision trees:

**IHSG(t+h) = (1/B) Ã— Î£áµ¢â‚Œâ‚á´® Táµ¢(X(t))**

Where:
- **Táµ¢(X(t))** = Prediction from i-th decision tree
- **h** = Forecast horizon (1 to 30 days)
- **B** = Number of trees (100)

### Multi-step Forecasting (Autoregressive)
For horizons > 1, the model uses its own predictions:

```
IHSG(t+1) = RF(X(t))
IHSG(t+2) = RF([IHSG(t+1), IHSG(t), ..., IHSG(t-18), STI(t+1), STI(t), ..., XAUUSD(t-18)])
...
IHSG(t+30) = RF([IHSG(t+29), ..., IHSG(t+10), STI(t+29), ..., XAUUSD(t+10)])
```

## Detailed Parameter Explanation

### 1. **lags=20**
- **Meaning**: Uses past 20 time steps of IHSG as features
- **Rationale**: 20 business days â‰ˆ 1 month of trading data
- **Technical**: Creates features IHSG(t-1), IHSG(t-2), ..., IHSG(t-20)

### 2. **lags_past_covariates=20** 
- **Meaning**: Uses past 20 time steps of STI and XAUUSD
- **Rationale**: Same temporal depth as target for consistency
- **Technical**: Creates 40 additional features (20 STI + 20 XAUUSD)

### 3. **output_chunk_length=1**
- **Meaning**: Predicts 1 day at a time
- **Alternative**: Could predict multiple days simultaneously
- **Advantage**: More stable for long horizons (30 days)

### 4. **n_estimators=100**
- **Meaning**: 100 decision trees in the forest
- **Trade-off**: More trees = better performance but slower training
- **Typical Range**: 50-500 trees

### 5. **max_depth=10**
- **Meaning**: Maximum depth of each decision tree
- **Purpose**: Prevents overfitting
- **Alternative**: Could use min_samples_leaf instead

### 6. **multi_models=True**
- **Meaning**: Separate model for each forecast horizon
- **Alternative**: Single model for all horizons
- **Advantage**: Each horizon optimized independently

### 7. **Business Day Frequency ('B')**
- **Meaning**: Only business days (Mon-Fri)
- **Automatic**: Excludes weekends
- **Limitation**: Doesn't handle public holidays automatically

```
â†“

â†“

```

## Performance Metrics Explained

### 1. **MAPE (Mean Absolute Percentage Error)**
```
MAPE = (100/n) Ã— Î£|Actual - Predicted|/|Actual|
```
- **Unit**: Percentage
- **Interpretation**: Lower is better
- **Typical Good Value**: < 5% for daily stock prices

### 2. **MAE (Mean Absolute Error)**
```
MAE = (1/n) Ã— Î£|Actual - Predicted|
```
- **Unit**: Same as target (IHSG points)
- **Interpretation**: Average absolute error
- **Robust**: Less sensitive to outliers than MSE

### 3. **RMSE (Root Mean Square Error)**
```
RMSE = âˆš[(1/n) Ã— Î£(Actual - Predicted)Â²]
```
- **Unit**: Same as target
- **Interpretation**: Penalizes large errors more
- **Comparison**: RMSE > MAE indicates some large errors

### 4. **Directional Accuracy**
```
DA = (1/n) Ã— Î£[sign(Actual_change) == sign(Predicted_change)]
```
- **Unit**: Percentage
- **Interpretation**: % of correct up/down predictions
- **Financial Relevance**: Important for trading decisions

---

## Model Assumptions & Limitations

### âœ… **Assumptions**
1. **Stationarity**: Market relationships remain relatively stable
2. **Covariate Availability**: STI and XAUUSD data available for forecast period
3. **Linear Relationships**: Random Forest can capture non-linear but still has limits
4. **No Structural Breaks**: Model assumes no major market regime changes

### âš ï¸ **Limitations**
1. **Horizon Dependency**: Accuracy typically decreases with longer horizons
2. **Market Volatility**: Extreme market events may not be well predicted
3. **Covariate Assumption**: Assumes perfect knowledge of STI/XAUUSD future values
4. **Static Model**: Doesn't adapt to changing market conditions without retraining

````python
# ðŸ†š RANDOM FOREST vs XGBOOST MODEL COMPARISON

print("="*80)
print("ðŸ†š COMPREHENSIVE MODEL COMPARISON")
print("="*80)

if 'xgb_forecast' in locals() and 'forecast' in locals():
    
    # Calculate performance metrics for both models
    print(f"\nðŸ“Š PERFORMANCE METRICS COMPARISON:")
    print("-" * 60)
    
    # Ensure both forecasts have same length for fair comparison
    min_length = min(len(forecast), len(xgb_forecast), len(ihsg_test))
    
    # Slice to common length
    rf_forecast_eval = forecast[:min_length]
    xgb_forecast_eval = xgb_forecast[:min_length]
    actual_eval = ihsg_test[:min_length]
    
    # Calculate metrics for Random Forest
    rf_mape = mape(actual_eval, rf_forecast_eval)
    rf_mae = mae(actual_eval, rf_forecast_eval)
    rf_rmse = rmse(actual_eval, rf_forecast_eval)
    
    # Calculate metrics for XGBoost
    xgb_mape = mape(actual_eval, xgb_forecast_eval)
    xgb_mae = mae(actual_eval, xgb_forecast_eval)
    xgb_rmse = rmse(actual_eval, xgb_forecast_eval)
    
    # Calculate directional accuracy
    actual_changes = actual_eval.diff().values()[1:]
    rf_changes = rf_forecast_eval.diff().values()[1:]
    xgb_changes = xgb_forecast_eval.diff().values()[1:]
    
    rf_direction_acc = (np.sign(actual_changes.flatten()) == np.sign(rf_changes.flatten())).mean() * 100
    xgb_direction_acc = (np.sign(actual_changes.flatten()) == np.sign(xgb_changes.flatten())).mean() * 100
    
    # Display comparison table
    comparison_data = {
        'Metric': ['MAPE (%)', 'MAE', 'RMSE', 'Directional Accuracy (%)'],
        'Random Forest': [f"{rf_mape:.2f}", f"{rf_mae:.2f}", f"{rf_rmse:.2f}", f"{rf_direction_acc:.2f}"],
        'XGBoost': [f"{xgb_mape:.2f}", f"{xgb_mae:.2f}", f"{xgb_rmse:.2f}", f"{xgb_direction_acc:.2f}"],
        'Winner': [
            'ðŸ† Random Forest' if rf_mape < xgb_mape else 'ðŸ† XGBoost',
            'ðŸ† Random Forest' if rf_mae < xgb_mae else 'ðŸ† XGBoost',
            'ðŸ† Random Forest' if rf_rmse < xgb_rmse else 'ðŸ† XGBoost',
            'ðŸ† Random Forest' if rf_direction_acc > xgb_direction_acc else 'ðŸ† XGBoost'
        ]
    }
    
    comparison_df = pd.DataFrame(comparison_data)
    print(comparison_df.to_string(index=False))
    
    # Calculate improvement percentages
    mape_improvement = ((rf_mape - xgb_mape) / rf_mape) * 100 if rf_mape != 0 else 0
    mae_improvement = ((rf_mae - xgb_mae) / rf_mae) * 100 if rf_mae != 0 else 0
    rmse_improvement = ((rf_rmse - xgb_rmse) / rf_rmse) * 100 if rf_rmse != 0 else 0
    
    print(f"\nðŸ“ˆ PERFORMANCE IMPROVEMENT:")
    print("-" * 40)
    if xgb_mape < rf_mape:
        print(f"âœ… XGBoost MAPE is {abs(mape_improvement):.2f}% better than Random Forest")
    else:
        print(f"âŒ Random Forest MAPE is {abs(mape_improvement):.2f}% better than XGBoost")
        
    if xgb_mae < rf_mae:
        print(f"âœ… XGBoost MAE is {abs(mae_improvement):.2f}% better than Random Forest")
    else:
        print(f"âŒ Random Forest MAE is {abs(mae_improvement):.2f}% better than XGBoost")
        
    if xgb_rmse < rf_rmse:
        print(f"âœ… XGBoost RMSE is {abs(rmse_improvement):.2f}% better than Random Forest")
    else:
        print(f"âŒ Random Forest RMSE is {abs(rmse_improvement):.2f}% better than XGBoost")
    
    # Create comprehensive visualization
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('Random Forest vs XGBoost Model Comparison', fontsize=16, fontweight='bold')
    
    # 1. Forecast Comparison
    ax1 = axes[0, 0]
    train_plot_start = max(0, len(ihsg_train) - 60)  # Last 60 days of training
    ihsg_train_plot = ihsg_train[train_plot_start:]
    
    ihsg_train_plot.plot(ax=ax1, label='Training Data', alpha=0.7, color='blue')
    rf_forecast_eval.plot(ax=ax1, label='Random Forest', color='red', linewidth=2)
    xgb_forecast_eval.plot(ax=ax1, label='XGBoost', color='green', linewidth=2)
    actual_eval.plot(ax=ax1, label='Actual', color='black', linestyle='--', alpha=0.8)
    
    ax1.set_title('Forecast Comparison')
    ax1.set_ylabel('IHSG Close Price')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Residuals Comparison
    ax2 = axes[0, 1]
    rf_residuals = actual_eval.values().flatten() - rf_forecast_eval.values().flatten()
    xgb_residuals = actual_eval.values().flatten() - xgb_forecast_eval.values().flatten()
    
    ax2.scatter(rf_forecast_eval.values().flatten(), rf_residuals, alpha=0.6, label='Random Forest', color='red')
    ax2.scatter(xgb_forecast_eval.values().flatten(), xgb_residuals, alpha=0.6, label='XGBoost', color='green')
    ax2.axhline(y=0, color='black', linestyle='--', alpha=0.8)
    ax2.set_xlabel('Predicted Values')
    ax2.set_ylabel('Residuals')
    ax2.set_title('Residuals vs Predicted')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Error Distribution
    ax3 = axes[0, 2]
    ax3.hist(rf_residuals, bins=20, alpha=0.6, label='Random Forest', color='red', density=True)
    ax3.hist(xgb_residuals, bins=20, alpha=0.6, label='XGBoost', color='green', density=True)
    ax3.set_xlabel('Residuals')
    ax3.set_ylabel('Density')
    ax3.set_title('Residuals Distribution')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 4. Actual vs Predicted Scatter
    ax4 = axes[1, 0]
    actual_values = actual_eval.values().flatten()
    
    ax4.scatter(actual_values, rf_forecast_eval.values().flatten(), alpha=0.6, label='Random Forest', color='red')
    ax4.scatter(actual_values, xgb_forecast_eval.values().flatten(), alpha=0.6, label='XGBoost', color='green')
    ax4.plot([actual_values.min(), actual_values.max()], [actual_values.min(), actual_values.max()], 'k--', linewidth=2)
    ax4.set_xlabel('Actual Values')
    ax4.set_ylabel('Predicted Values')
    ax4.set_title('Actual vs Predicted')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    # 5. Performance Metrics Bar Chart
    ax5 = axes[1, 1]
    metrics = ['MAPE', 'MAE', 'RMSE']
    rf_values = [rf_mape, rf_mae, rf_rmse]
    xgb_values = [xgb_mape, xgb_mae, xgb_rmse]
    
    x = np.arange(len(metrics))
    width = 0.35
    
    ax5.bar(x - width/2, rf_values, width, label='Random Forest', color='red', alpha=0.7)
    ax5.bar(x + width/2, xgb_values, width, label='XGBoost', color='green', alpha=0.7)
    
    ax5.set_ylabel('Error Value')
    ax5.set_title('Performance Metrics Comparison')
    ax5.set_xticks(x)
    ax5.set_xticklabels(metrics)
    ax5.legend()
    ax5.grid(True, alpha=0.3)
    
    # 6. Cumulative Error
    ax6 = axes[1, 2]
    rf_cumulative_error = np.cumsum(np.abs(rf_residuals))
    xgb_cumulative_error = np.cumsum(np.abs(xgb_residuals))
    
    ax6.plot(range(len(rf_cumulative_error)), rf_cumulative_error, label='Random Forest', color='red', linewidth=2)
    ax6.plot(range(len(xgb_cumulative_error)), xgb_cumulative_error, label='XGBoost', color='green', linewidth=2)
    ax6.set_xlabel('Time Steps')
    ax6.set_ylabel('Cumulative Absolute Error')
    ax6.set_title('Cumulative Error Comparison')
    ax6.legend()
    ax6.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Feature importance comparison (if available)
    print(f"\nðŸ” FEATURE IMPORTANCE ANALYSIS:")
    print("-" * 50)
    
    try:
        # Get Random Forest feature importance
        rf_estimator = rf_model.get_estimator(horizon=1, target_dim=0)
        rf_feature_names = rf_model.lagged_feature_names
        
        if hasattr(rf_estimator, 'feature_importances_'):
            rf_importance = rf_estimator.feature_importances_
            
            # Get XGBoost feature importance
            xgb_estimator = xgb_model.get_estimator(horizon=1, target_dim=0)
            
            if hasattr(xgb_estimator, 'feature_importances_'):
                xgb_importance = xgb_estimator.feature_importances_
                
                # Create comparison dataframe
                importance_comparison = pd.DataFrame({
                    'Feature': rf_feature_names,
                    'Random_Forest': rf_importance,
                    'XGBoost': xgb_importance
                })
                
                # Calculate correlation between feature importances
                importance_corr = np.corrcoef(rf_importance, xgb_importance)[0, 1]
                
                print(f"Feature Importance Correlation: {importance_corr:.4f}")
                print("(Higher correlation = models agree on important features)")
                
                # Show top 10 features for each model
                print(f"\nTop 10 Features - Random Forest:")
                rf_top = importance_comparison.nlargest(10, 'Random_Forest')[['Feature', 'Random_Forest']]
                print(rf_top.to_string(index=False))
                
                print(f"\nTop 10 Features - XGBoost:")
                xgb_top = importance_comparison.nlargest(10, 'XGBoost')[['Feature', 'XGBoost']]
                print(xgb_top.to_string(index=False))
                
            else:
                print("XGBoost feature importance not available")
        else:
            print("Random Forest feature importance not available")
            
    except Exception as e:
        print(f"Could not extract feature importance: {e}")
    
    # Statistical significance test
    print(f"\nðŸ“Š STATISTICAL SIGNIFICANCE TEST:")
    print("-" * 50)
    
    from scipy.stats import ttest_rel
    
    # Paired t-test for residuals
    try:
        rf_abs_errors = np.abs(rf_residuals)
        xgb_abs_errors = np.abs(xgb_residuals)
        
        t_stat, p_value = ttest_rel(rf_abs_errors, xgb_abs_errors)
        
        print(f"Paired t-test (absolute errors):")
        print(f"  â€¢ t-statistic: {t_stat:.4f}")
        print(f"  â€¢ p-value: {p_value:.6f}")
        print(f"  â€¢ Result: {'Statistically significant difference' if p_value < 0.05 else 'No significant difference'}")
        
        if p_value < 0.05:
            better_model = "XGBoost" if xgb_abs_errors.mean() < rf_abs_errors.mean() else "Random Forest"
            print(f"  â€¢ Winner: {better_model} is significantly better")
            
    except Exception as e:
        print(f"Could not perform statistical test: {e}")
    
    # Summary and recommendations
    print(f"\nðŸŽ¯ SUMMARY & RECOMMENDATIONS:")
    print("-" * 50)
    
    # Determine overall winner
    rf_wins = sum([rf_mape < xgb_mape, rf_mae < xgb_mae, rf_rmse < xgb_rmse, rf_direction_acc > xgb_direction_acc])
    xgb_wins = 4 - rf_wins
    
    if xgb_wins > rf_wins:
        print("ðŸ† OVERALL WINNER: XGBoost")
        print(f"   â€¢ XGBoost outperforms Random Forest on {xgb_wins}/4 metrics")
        print("   â€¢ Recommended for final IHSG forecasting model")
    elif rf_wins > xgb_wins:
        print("ðŸ† OVERALL WINNER: Random Forest")
        print(f"   â€¢ Random Forest outperforms XGBoost on {rf_wins}/4 metrics")
        print("   â€¢ Recommended for final IHSG forecasting model")
    else:
        print("ðŸ¤ TIE: Both models perform similarly")
        print("   â€¢ Consider ensemble approach or model averaging")
    
    print(f"\nðŸ’¡ THESIS IMPLICATIONS:")
    print("   â€¢ Both models demonstrate strong predictive capability")
    print("   â€¢ Model comparison validates robustness of results")
    print("   â€¢ Gradient boosting vs ensemble methods analysis")
    print("   â€¢ Feature importance provides insight into market dynamics")
    
else:
    print("âŒ Models not available. Please run Random Forest and XGBoost training first.")

print("="*80)
````
